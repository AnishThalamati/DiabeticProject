##### cgm data manipulation
cgm is the original data set
final_cgm is the data set without the Scan_Glucose column
mean_cgm is the mean of "Historic_ Glucose" by date
meancgm is the data with "Historic_ Glucose" and "cgmdate" without the "cgmtime"

```python
cgm.head()
```

```python
cgm.loc[cgm["Historic_ Glucose"].isnull(), "Historic_ Glucose"] = cgm["Scan_Glucose"]
```

```python
cgm.head()
```

```python
final_cgm = cgm.loc[:,["cgmdate", "cgmtime","Historic_ Glucose"]]
```

##### grouping by each day

```python
final_cgm.groupby(['cgmdate','cgmtime']).sum()
```

```python
mean_cgm = final_cgm.groupby(['cgmdate'], as_index = False).mean() 
```

```python
mean_cgm.head()
```

```python
meancgm = mean_cgm.loc[:,['cgmdate','Historic_ Glucose']]
```

```python
meancgm.head(1)
```

```python
final_cgm.groupby(["cgmdate","Historic_ Glucose" ],as_index = False).mean()
```

##### manipultaion of steps  and gps data
final_steps is the dataset after removing the unnecessary columns

cgm_ steps = merged meancgm and steps data as a single dataset

cgm_gps = merged meancgm and gps data as a single dataset

```python
steps.head(1)
```

```python
final_steps = steps.loc[:,["stepsdate","active_time","distance","steps"]]
```

```python
final_steps.shape
```

```python
cgm_steps = meancgm.merge(final_steps, how='inner',left_on = 'cgmdate', right_on = 'stepsdate')
```

```python
cgm_steps.head()
```

```python
gps.head(2)
```

```python
cgm_gps = meancgm.merge(gps, how = 'inner', left_on ='cgmdate', right_on = 'startdate' )
```

```python
cgm_gps.head(2)
```

##### merging cgm_steps and cgm_gps as a single dataset
this data set is named as cgm_steps_gps

```python
cgm_steps_gps = cgm_gps.merge(cgm_steps, how = 'inner', left_on = 'cgmdate', right_on = 'cgmdate')
```

```python
cgm_steps_gps.head(7)
```
