Creating pair plots with the assessed features tested against 'Historic_ Glucose' 

```python
glucose_date = sns.pairplot(df, x_vars = ['cgmdate']  , y_vars = 'Historic_ Glucose',size = 10, aspect =0.7, kind= 'reg')
```

    3. Few days had prominent spikes of the glucose levels

```python
sns.pairplot(df, y_vars = 'Historic_ Glucose', x_vars = ['steps'], size = 5, kind = 'reg')
```

```python
sns.barplot(x = 'steps', y = 'Historic_ Glucose', data = df)
```

Insights: 

    1. The intercept falls at approximately 140  which means that even if the patient had '0' activity his glucose level remained normal.
    
    2. Even with the maximum activity, his blood glucose levels were high. 
    
    3. Amount of activity might not play a major role in maintaining blood glucose levels
    
    4. Glucose levels are not in direct relation to the number of steps taken by the user. 
    
    5. Contrast to the claims of the device manufacturer that it can be used as a method to control glucose in diabetics.

```python
sns.pairplot(df, y_vars = 'Historic_ Glucose', x_vars = ['sleep_duration'], size = 5, kind = 'reg')
```

```python
sns.barplot(x = 'sleep_duration', y = 'Historic_ Glucose', data = df)
```

Insights: 

    1. The intercept falls at approximately 180  which means that if the patient had '0' sleep his glucose level increases.
    
    2. Increasing the duration of sleep can better control the blood glucose levels. 
    
    3. Amount of sleep might play a major role in maintaining blood glucose levels than the amount of activity.
